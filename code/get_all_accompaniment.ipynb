```python
import os
import shutil

target_path = os.getcwd()+'/spleeter_output'
new_path = './all_accompaniment/'
for root,dirs,files in os.walk(target_path):
    for file in files:
        #获取文件所属目录
        print(root)
        #获取文件路径
        print(os.path.join(root,file))
        if 'accompaniment' in file:
            new_file_name = root.replace(target_path+'\\','') + '.wav'
            shutil.copy(os.path.join(root,file), new_path + new_file_name) 
            print('copy ',os.path.join(root,file),' to ',new_path + new_file_name)

            
```
